```python
qna=[]
for page in reader.pages:
    text=page.extract_text()
    print('******')
    lines=text.split('\n')
    for i in range(len(lines)):
        
        line=lines[i]
        next_line=lines[i+1] if i+1<len(lines) else ''
        if line.startswith('Q:'):
            qna.append({'Question':line})
            if next_line.startswith('A:'):
                qna[-1]['Answer']=next_line
            else:
                qna[-1]['Answer']='N/A'
            
```

```python
from sentence_transformers import SentenceTransformer, util
# model = SentenceTransformer('bert-base-nli-mean-tokens')
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')
import torch
```

```python
embedded_questions=[torch.tensor(model.encode(set['Question'])) for set in qna]
embedded_answers=[torch.tensor(model.encode(set['Answer'])) for set in qna]
```

```python
# Function to compare two strings using semantic similarity
def compare_strings(string1_embedding, string2_embedding):
    similarity = util.pytorch_cos_sim(string1_embedding, string2_embedding)[0][0]
    return similarity

# Example usage
ques_score=[]
ans_score=[]
user_input = input('Ask a question: ')
input_embedding = model.encode([user_input])
for i in range(len(embedded_questions)):
    ques_score.append(compare_strings(input_embedding,embedded_questions[i]))
    ans_score.append(compare_strings(input_embedding,embedded_answers[i]))
    
max_ques=max(ques_score)
response=qna[ques_score.index(max(ques_score))]['Answer'] if max(ques_score)>max(ans_score) else qna[ans_score.index(max(ans_score))]['Answer']
print(response[3:])
```

```python
from sentence_transformers import SentenceTransformer, util
import torch

# Load a pre-trained BERT-based model
model = SentenceTransformer('bert-base-nli-mean-tokens')

# Function to compare two strings using semantic similarity
def compare_strings(string1, string2):
    string1_embedding = model.encode([string1])[0]
    string2_embedding = model.encode([string2])[0]

    # Convert embeddings to torch tensors
    string1_embedding = torch.tensor(string1_embedding)
    string2_embedding = torch.tensor(string2_embedding)

    # Calculate cosine similarity between the tensors
    similarity = util.pytorch_cos_sim(string1_embedding, string2_embedding)

    return similarity.item()

# Example usage
string1 = "height of Mount Everest"
string2 = "What is the height of Mount Everest?"

similarity_score = compare_strings(string1, string2)
print("Similarity score:", similarity_score)
```
